```python
import pandas as pd 
import numpy as np
import os, sys
import pickle
```

```python
#chanpinName = ['chanpinjingli1.data','chanpinbujingli.data','dianshangchanpinjingli.data','chanpinzhuli.data','chanpinzongjian.data','shujuchanpinjingli.data']
chanpinName = []
for i in os.listdir(os.getcwd()):
    if '.data' in i:
        chanpinName.append(i)
```

```python
len(chanpinName)
```

```python
dfList = [pd.read_csv(i,header= None,sep='|',error_bad_lines =False) for i in chanpinName]
#chanpinDF = pd.concat(chanpinData)
df = pd.concat(dfList)
```

```python
df
```

```python
count = df[[0,1]].groupby(1).count().sort_values(0)
idx = count[count.values>100].index
iidx = [i in idx for i in df.values[:,1]]
```

```python
washData = df[iidx].reset_index()[[1,2,3,4]]
```

```python
minS = []
maxS = []
for i in washData[2].values:
    if i[-2:]=='以上':
        #print(i)
        f = i.lower().split('k')
        minS.append(int(f[0])*1000)
        maxS.append(int(f[0])*1000)
        continue
    f = i.lower().split('-')
    minS.append(int(f[0][:-1])*1000)
    maxS.append(int(f[1][:-1])*1000)
```

```python
washData[5] = minS
washData[6] = maxS
washData = washData[[1,3,4,5,6]]
```

```python
#def fun(i,name)
i =1
ser = washData[i]
s= ser.unique()
s.sort()
m = dict(zip(s,range(len(s))))
washData[i] = washData[i].map(m)
with open('dict/occupation'+'.pkl','wb') as outfile:
    pickle.dump(m,outfile)
   # json.dump(m,outfile,ensure_ascii=False)  
   # outfile.write('\n')

i =3
ser = washData[i]
s= ser.unique()
s.sort()
m = dict(zip(s,range(len(s))))
washData[i] = washData[i].map(m)
with open('dict/city'+'.pkl','wb') as outfile:  
    pickle.dump(m,outfile)
   # json.dump(m,outfile,ensure_ascii=False)  
   # outfile.write('\n')
i =4
ser = washData[i]
s= ser.unique()
s.sort()
m = dict(zip(s,range(len(s))))
washData[i] = washData[i].map(m)
with open('dict/experience'+'.pkl','wb') as outfile:  
    pickle.dump(m,outfile)
   # json.dump(m,outfile,ensure_ascii=False)  
   # outfile.write('\n')
```

```python
train  = washData.sample(frac=1.0)
trainIdx = train.index
testIdx = washData.index.difference(trainIdx)
test = washData.loc[testIdx]
```

```python
train.to_csv('train.csv',header=None,index=None)
test.to_csv('test.csv',header=None,index=None)
```
